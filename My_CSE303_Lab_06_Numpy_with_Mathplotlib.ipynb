```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6), dpi=80)


X = np.linspace(-2*np.pi, 2*np.pi, 500,endpoint=True)
C = np.cos(X)
S = np.sin(X)
T = np.tan(X)

#print(C)
#print(S)

plt.subplot(3,1,1)

plt.title("Cosine Graph")
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],[r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1, 0, +1],[r'$-1$', r'$0$', r'$+1$'])

plt.plot(X, C, color="red", label="cosine", linewidth=1.0, linestyle="-")

t = 0
plt.plot([t,t], [-1,np.cos(t)],color = 'red', linestyle = "--", linewidth = 1.0)
plt.annotate(r'$\cos(\frac{2\pi}{3})=-\frac{1}{2}$',xy=(t, np.cos(t)), xycoords='data',xytext=(-90, -50), textcoords='offset points', fontsize=16,arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))

plt.subplot(3,1,2)


plt.title("Sine Graph")
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],[r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1, 0, +1],[r'$-1$', r'$0$', r'$+1$'])

plt.plot(X, S, color="blue", label="sine", linewidth=1.0, linestyle="-")

plt.subplot(3,1,3)




plt.plot(X, T, color="green", label="tan", linewidth=1.0, linestyle="-")

plt.xlim(-4.0,4.0)
#plt.ylim(-1.0,1.0)

#plt.xticks(np.linspace(-4,4,9,endpoint=True))
#plt.yticks(np.linspace(-1,1,5,endpoint=True))

#plt.xticks(np.arange(-4,4,1))
#plt.yticks(np.arange(-1,1.5,.5))

plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],[r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1, 0, +1],[r'$-1$', r'$0$', r'$+1$'])

plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.legend(loc = "upper left")

plt.title("Tan Graph")

plt.savefig("sample1.png",dpi=72)

plt.subplots_adjust(hspace = 1)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure (figsize=(8,6), dpi = 80)

X = np.linspace(-2*np.pi, 2*np.pi, 500, endpoint = True)
C = np.cos(X)
S = np.sin(X)
T = np.tan(X)

plt.subplot(3,1,1)
plt.plot(X, C, color = "red", label = "cosine", linestyle = "-", linewidth = 2.0)
plt.xticks([-2*np.pi, -1.5*np.pi, -np.pi, -np.pi/2, 0, np.pi/2, np.pi, 1.5*np.pi, 2*np.pi],
           [r'$-2\pi$', r'$-3\pi/2$', r'$-\pi$', r'$-\pi/2$', r'$0$', r'$\pi/2$', r'$\pi$', r'$3\pi/2$', r'$-2\pi$'])

t = 0
plt.plot([t,t], [-1, np.cos(t)], color = 'red', linestyle = "--", linewidth = 1.0)
plt.annotate(r'$ \cos0 = 1$', xy = (t, np.cos(t)), xycoords = 'data', xytext = (+100, -30), textcoords = 'offset points',
             fontsize = 10, arrowprops = dict(arrowstyle= '->', connectionstyle = "arc3, rad = .2"))
plt.title("Cosine Curve")

plt.subplot(3,1,2)
plt.plot(X, S, color = "blue", label = "sine", linestyle = "-", linewidth = 2.0)
plt.xticks([-2*np.pi, -1.5*np.pi, -np.pi, -np.pi/2, 0, np.pi/2, np.pi, 1.5*np.pi, 2*np.pi],
           [r'$-2\pi$', r'$-3\pi/2$', r'$-\pi$', r'$-\pi/2$', r'$0$', r'$\pi/2$', r'$\pi$', r'$3\pi/2$', r'$-2\pi$'])
plt.title("Sine Curve")

plt.subplot(3,1,3)
plt.plot(X, T, color = "green", label = "tan", linestyle = "-", linewidth = 2.0)

plt.xlim(-4.0, 4.0)
#plt.ylim(-1.0, 1.0)

#plt.xticks (np.arange(-4,5,1))
plt.xticks([-2*np.pi, -1.5*np.pi, -np.pi, -np.pi/2, 0, np.pi/2, np.pi, 1.5*np.pi, 2*np.pi],
           [r'$-2\pi$', r'$-3\pi/2$', r'$-\pi$', r'$-\pi/2$', r'$0$', r'$\pi/2$', r'$\pi$', r'$3\pi/2$', r'$-2\pi$'])
#plt.yticks (np.arange(-1,1.5,0.5))

plt.xlabel("x-axis")
plt.ylabel("y-axis")

#plt.legend(loc = "upper left")

plt.title("Tan Curve")
plt.savefig("sample1.png",dpi=72)

plt.subplots_adjust(hspace = 1)
plt.show()
```
