```python
import statsmodels.api as sm
import pandas as pd
```

```python
drop_dc=df.drop(['Day','Lab Test','Confirmed case'],  axis=1).values
drop_dc
```

```python
mean = drop_dc.mean()
mean
```

```python
n = drop_dc.shape[0]
n
```

```python
drop_dc[1][0]
```

```python
target = []
for i in range (n):
  if drop_dc [i][0]<=mean:
    target.append(0)
  else:
    target.append(1)

print(target)



```

```python
target = pd.DataFrame(target)
target
```

```python
target = target.set_axis(['Rate'], axis=1)
target
```

```python
df['target']=target
df
```

```python
x=df.drop(['Day','Death Case','target'], axis=1).values
y=df.drop(['Day','Death Case','Lab Test','Confirmed case'],  axis=1).values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
mdl = LinearRegression()
mdl.fit(X_train, Y_train)
```

```python
y_pred=mdl.predict(X_test)
print(y_pred)
```

```python
import statsmodels.api as sm
import pandas as pd
```

```python
logistic = sm.Logit(Y_train, X_train).fit()
```

```python
logistic.summary()
```

```python
Y1_pred = logistic.predict(X_test)
print(Y1_pred)
```

```python
prediction = list(map(round,Y1_pred))
print(prediction)
```

```python
print('Actual class: ', list(Y_test))
print('Predicted class: ', prediction)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score

cm = confusion_matrix (Y_test, prediction)
print(cm)

print(accuracy_score(Y_test, prediction))
```

```python
import numpy as np
from sklearn.linear_model import LogisticRegression
```

```python
logistic_reg = LogisticRegression(C = 0.1)
logistic_reg.fit(X_train, Y_train)
```

```python
prediction_log_reg = logistic_reg.predict(X_test)
print('Actual Class:                       ', np.array(Y_test))
print('Predicted class(using sklearn):     ',  prediction_log_reg)
print('Predicted class(using statsmodel):  ', np.array(prediction))
```

```python
cm1 = confusion_matrix (Y_test, prediction_log_reg)
print(cm1)

print(accuracy_score(Y_test, prediction_log_reg))
```

```python
print(logistic_reg.intercept_)
print(logistic_reg.coef_)
```

***********************************SVM

```python
from sklearn import svm

model = svm.SVC(kernel = 'linear', C = 1, probability = True )
```

```python
model.fit(X_train, Y_train)
```

```python
Y_pred = model.predict(X_test)
```

```python
compare_dataset = pd.DataFrame({'Actual': Y_test.flatten(), 'Predicted': Y_pred.flatten()})
```

```python
print(compare_dataset)
```

```python
from sklearn import metrics
print("Accuracy: ", metrics.accuracy_score(Y_test, Y_pred))
print("Precision: ", metrics.precision_score(Y_test, Y_pred))
print("Recall: ", metrics.recall_score(Y_test, Y_pred))
```

```python
y_pred_train = model.predict(X_train)

print("Training Accuracy: ", metrics.accuracy_score(Y_train, y_pred_train))
```

```python
cm = metrics.confusion_matrix (Y_test, Y_pred)
print(cm)
```

```python
from sklearn.metrics import roc_curve, auc

logistic_fpr, logistic_tpr, threshold = roc_curve(Y_test, Y1_pred)
auc_logistic = auc(logistic_fpr, logistic_tpr)

svm_fpr, svm_tpr, threshold = roc_curve(Y_test, Y_pred)
auc_svm = auc(svm_fpr, svm_tpr)

plt.figure(figsize=(8, 8), dpi=100)
plt.plot(svm_fpr, svm_tpr, linestyle='-', label='SVM (auc = %0.3f)' % auc_svm)
plt.plot(logistic_fpr, logistic_tpr, marker='.', label='Logistic (auc = %0.3f)' % auc_logistic)

plt.xlabel('False Positive Rate -->')
plt.ylabel('True Positive Rate -->')

plt.legend()

plt.show()
```
