```python
temp = df.groupby('Day')['Lab Test', 'Confirmed case', 'Death Case'].sum().reset_index()
temp = temp.melt(id_vars='Day', value_vars=['Lab Test', 'Confirmed case', 'Death Case'], var_name='Case', value_name='Count')

fig = px.area(temp, x = 'Day', y = 'Count', color= 'Case', height = 600, title='Cases Over Time', color_discrete_sequence=[labt, cnf, dth] )
fig.update_layout(xaxis_rangeslider_visible = True)
fig.show()
```

```python
fig_c = px.bar(df, x = 'Day', y = 'Confirmed case', color_discrete_sequence=[cnf])
fig_d = px.bar(df, x = 'Day', y = 'Death Case', color_discrete_sequence=[dth])

fig = make_subplots(rows=1 , cols=2, shared_xaxes=False, horizontal_spacing=0.1, subplot_titles=('Confirmed Cases', 'Death Cases'))

fig.add_trace(fig_c['data'][0], row=1, col=1)
fig.add_trace(fig_d['data'][0], row=1, col=2)

fig.update_layout(height = 500)
fig.show()
```

```python
fig1 = px.line(df, x = 'Day', y = 'Lab Test', color_discrete_sequence=[labt])
fig2 = px.line(df, x = 'Day', y = 'Confirmed case', color_discrete_sequence=[cnf])
fig3 = px.line(df, x = 'Day', y = 'Death Case', color_discrete_sequence=[dth])

fig = make_subplots(rows=1 , cols=3, shared_xaxes=False, horizontal_spacing=0.1, subplot_titles=('Lab Test', 'Confirmed case', 'Death Case'))

fig.add_trace(fig1['data'][0], row=1, col=1)
fig.add_trace(fig2['data'][0], row=1, col=2)
fig.add_trace(fig3['data'][0], row=1, col=3)

fig.update_layout(height = 500)
fig.show()
```
